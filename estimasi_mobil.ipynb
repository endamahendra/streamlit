# Jumlah mobil berdasarkan model


```python
models = df.groupby('model').count()[['price']].sort_values(by='price',ascending=True).reset_index()
models = models.rename(columns={'price' : 'Jumlah Mobil'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=models['model'], y=models['Jumlah Mobil'], color='royalblue')
plt.xticks(rotation=60)
```

# Jumlah mobil berdasarkan ukuran mesin


```python
engine = df.groupby('engineSize').count()[['tax']].sort_values(by='tax',ascending=True).reset_index()
engine = engine.rename(columns={'tax' : 'count'})
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x=engine['engineSize'], y=engine['count'], color='royalblue')
```

# seleksi fitur


```python
features = ['year', 'mileage', 'tax', 'mpg', 'engineSize']
X = df[features]
y = df['price']
X.shape, y.shape
```

# split data training dan data testing


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,  random_state=70)
y_train.shape, y_test.shape
```

# membuat model regresi linier


```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

# membuat inputan model regresi linier


```python
#year=2019, mileage=5000, tax=145, mpg=30.2, engineSize=2.0
input_data = np.array([2019, 5000, 145, 30.2, 2.0])
input_data = input_data.reshape(1, -1)
prediction = lr.predict(input_data)
print('Estimasi Harga mobil dalam EUR : ', prediction)
```

# Export model


```python
import pickle
filename = 'estimasi_mobil.sav'
pickle.dump(lr, open(filename, 'wb'))
```
